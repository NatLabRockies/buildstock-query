```python
from pathlib import Path
folder_path = Path("comstock_oedi_state_and_county/upgrade=0")
```

```python
# the full set is too big to load all into memory - use polars to scan and collect only relevant columns
df = pl.scan_parquet(folder_path, hidden_file_prefix=".")
reuse_rate = (
    df.group_by("in.state")
    .agg([
        pl.len().alias("total_rows"),
        pl.col("bldg_id").n_unique().alias("distinct_bldgs")
    ])
    .with_columns(
        (pl.col("total_rows") / pl.col("distinct_bldgs")).round(2).alias("reuse_rate")
    )
    .sort("reuse_rate", descending=True)
    .collect()
)
reuse_rate
```

```python
reuse_rate_agg = reuse_rate.select(
    pl.col("total_rows").sum().alias("total_rows"),
    pl.col("distinct_bldgs").sum().alias("distinct_bldgs"),
    pl.col("reuse_rate").mean().alias("reuse_rate_mean"),
)
reuse_rate_agg
```
